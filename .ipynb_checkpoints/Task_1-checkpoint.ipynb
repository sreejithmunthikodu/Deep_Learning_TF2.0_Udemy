## Simple Linear Regression

### Import Packages

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import pdb
```

### Generate random data

```python
observations = 1000
x1 = np.random.uniform(-10, 10, observations)
x2 = np.random.uniform(-10, 10, observations)

x = np.column_stack((x1, x2))
print(x.shape)
```

### Generate targets
y = 25(x1) + 300(x2) + 500 + noise 

```python
noise = np.random.uniform(-1, 1, observations)
y = 25*(x1) - 300*(x2) + 500 + noise
y = y.reshape(y.shape[0],1)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")
ax.scatter(x1,x2,y)
```

### Initialize variables

```python
init_range = 0.1
weights = np.random.uniform(-init_range, init_range, size=(2,1))
bias = np.random.uniform(-init_range, init_range, size=1)
weights
```

### Select a learning rate

```python
alpha = 0.02
```

### Train the model

```python
%%time
alpha = 0.02
init_range = 0.1
weights = np.random.uniform(-init_range, init_range, size=(2,1))
bias = np.random.uniform(-init_range, init_range, size=1)
weights

# set maximum number of iterations
max_iter = 1000
loss_history = []
for i in range(max_iter):
    y_pred = np.matmul(x, weights) + bias
    # loss function = 1/2 L2 Norm
    delta_y = (y_pred - y)
    loss = np.sum(delta_y**2)/2/observations

    loss_history.append(loss)
    delta_y = delta_y/observations
    # Update weights and biases
 
    delta_weights = np.matmul(x.T, delta_y)
    delta_bias = np.sum(delta_y)
    
    weights = weights - (alpha * delta_weights)
    bias = bias - (alpha * delta_bias)
  
```

```python
plt.plot(list(range(max_iter)), loss_history)
```
